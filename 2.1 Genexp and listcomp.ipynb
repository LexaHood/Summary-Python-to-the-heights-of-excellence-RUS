```python
# Списковое включение 
colors = ['Красные', 'Черные']
sizes = ['Пики', 'Черви', 'Бубны', 'Крести']
bodys = ['Туз', "Король", "Королева", "Валет", "10", "9", "8", "7", "6"]
tshirts = [(color, size, body) for color in colors for size in sizes for body in bodys]
tshirts
```

```python
# Генераторное выражние
# Использует меньше памяти т.к. отдает элементы по дному, вместо того чтобы строить список целиком и передавать его др конструктору
tshirts_2 = []
for color in colors: 
    for size in sizes: 
        for body in bodys:
            tshirts_2.append((color, size, body))
tshirts_2
```
